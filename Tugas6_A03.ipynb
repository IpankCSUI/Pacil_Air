## Data Preparation

### Handle Missing Value

```python
# Get started to do data preparation by checking null value in the dataset
df.isnull().sum(axis = 0)
```

```python
# Mengganti nilai yang hilang dengan nilai rata-rata
# Untuk variabel numerik
df['Flight Distance'].fillna(df['Flight Distance'].mean(), inplace=True)
df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean(), inplace=True)

# Untuk variabel kategoris, misalnya 'Gender' dan 'Type of Travel'
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)  # Menggunakan modus
df['Type of Travel'].fillna(df['Type of Travel'].mode()[0], inplace=True)  # Menggunakan modus
df['Class'].fillna(df['Class'].mode()[0], inplace=True)  # Menggunakan modus
df['Satisfaction'].fillna(df['Satisfaction'].mode()[0], inplace=True)  # Menggunakan modus


# Check again
df.isnull().sum(axis = 0)
```

### Handle Outlier

```python
# melakukan cek outliers dengan IQR
Q3 = df.quantile(0.75)
Q1 = df.quantile(0.25)
IQR = Q3-Q1
Jumlah_Outlier = []

for col in df.select_dtypes(np.number).columns:
    outliers = ((df[col] < (Q1[col] - 1.5 * IQR[col])) | (df[col] > (Q3[col] + 1.5 * IQR[col]))).sum()
    Jumlah_Outlier.append([col, outliers, outliers/len(df)])

indexOutput = list(range(0, len(Jumlah_Outlier)))

pd.DataFrame(Jumlah_Outlier, columns=['Column', 'Outlier','Persentase'],index=indexOutput).sort_values(by=['Persentase'], ascending = False,ignore_index=True)
```

```python
# Mengubah tipe data kolom 'id' menjadi object
df['id'] = df['id'].astype('object')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_numeric = df.select_dtypes(include=['number'])
# Membuat boxplot horizontal untuk setiap kolom numerik
for column in df_numeric.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column], orient='h')
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
def Outliers(df, column):
    if column == 'obj':
        return None

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_df = df.loc[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outlier_df

CT_Outliers = Outliers(df, 'Customer Type')
print("Customer Type Outliers:")
print(CT_Outliers['Customer Type'].unique())

DD_Outliers = Outliers(df, 'Departure Delay in Minutes')
print("\nDeparture Delay in Minutes Outliers:")
print(DD_Outliers['Departure Delay in Minutes'].unique())

CS_Outliers = Outliers(df, 'Checkin service')
print("\nCheckin service Outliers:")
print(CS_Outliers['Checkin service'].unique())

OS_Outliers = Outliers(df, 'On-board service')
print("\nOn-board service Outliers:")
print(OS_Outliers['On-board service'].unique())

FD_Outliers = Outliers(df, 'Flight Distance')
print("\nFlight Distance Outliers:")
print(FD_Outliers['Flight Distance'].unique())
```

```python
# Mengubah tipe data kolom 'id' menjadi object
df['id'] = df['id'].astype(str)
```

#### Apa Dampak Keterlambatan Penerbangan terhadap Kepuasan Penumpang?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Departure Delays
plt.figure(figsize=(10, 6))
sns.histplot(df['Departure Delay in Minutes'], bins=50, kde=True)
plt.title('Distribution of Departure Delays')
plt.xlabel('Minutes')
plt.ylabel('Frequency')
plt.show()

# Distribution of Arrival Delays
plt.figure(figsize=(10, 6))
sns.histplot(df['Arrival Delay in Minutes'], bins=50, kde=True)
plt.title('Distribution of Arrival Delays')
plt.xlabel('Minutes')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of Satisfaction by Delay Categories
df['Delayed'] = df['Arrival Delay in Minutes'] > 0

plt.figure(figsize=(20, 6))
sns.boxplot(x='Arrival Delay in Minutes', y='Satisfaction', data=df)
plt.title('Satisfaction by Delay Status')
plt.xlabel('Arrival Delay in Minutes')
plt.ylabel('Satisfaction')
plt.show()
```

```python
# Boxplot of Satisfaction by Delay Categories
df['DepartureDelayed'] = df['Departure Delay in Minutes'] > 0

plt.figure(figsize=(80, 6))
sns.boxplot(x='Departure Delay in Minutes', y='Satisfaction', data=df)
plt.title('Satisfaction by Delay Status')
plt.xlabel('Departure Delay in Minutes')
plt.ylabel('Satisfaction')
plt.show()
```

```python
from scipy.stats import ttest_ind

# Split data into delayed and non-delayed groups
delayed = df[df['Delayed'] == True]['Satisfaction']
not_delayed = df[df['Delayed'] == False]['Satisfaction']

# Perform t-test
t_stat, p_value = ttest_ind(delayed, not_delayed)

print(f'T-statistic: {t_stat}, P-value: {p_value}')
```

```python
# Split data into delayed and non-delayed groups
departuredelayed = df[df['Delayed'] == True]['Satisfaction']
not_departuredelayed = df[df['Delayed'] == False]['Satisfaction']

# Perform t-test
t_statdepdelayed, p_valuedepdelayed = ttest_ind(departuredelayed, not_departuredelayed)

print(f'T-statistic: {t_statdepdelayed}, P-value: {p_valuedepdelayed}')
```

```python
# Average Satisfaction by Delay Status
avg_satisfaction = df.groupby('Delayed')['Satisfaction'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Delayed', y='Satisfaction', data=avg_satisfaction)
plt.title('Average Satisfaction by Arrival Delay')
plt.xlabel('Delayed')
plt.ylabel('Average Satisfaction')
plt.show()
```

```python
avg_satisfaction = df.groupby('DepartureDelayed')['Satisfaction'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='DepartureDelayed', y='Satisfaction', data=avg_satisfaction)
plt.title('Average Satisfaction by Departure Delay')
plt.xlabel('DepartureDelayed')
plt.ylabel('Average Satisfaction')
plt.show()
```

Conclusion

- The distribution of departure and arrival delays shows that most flights have minimal delays, with a few having significant delays.
- Boxplots and statistical analysis indicate that delayed flights generally have lower satisfaction levels compared to non-delayed flights.
- The t-test results (T-statistic and P-value) confirm that the difference in satisfaction between delayed and non-delayed flights is statistically significant.
- On average, passengers on delayed flights report lower satisfaction scores, highlighting the importance of minimizing delays to improve overall passenger satisfaction.

These insights can help airlines prioritize efforts to reduce delays, thereby enhancing customer experience and satisfaction.


## Prediction

```python
# Prediction Model
from sklearn.metrics import precision_score, \
    recall_score, classification_report, \
    accuracy_score, f1_score

def evaluate_classifier_performance(prediction, y_test):
    # Informasi evaluasi secara compact
    print("Hasil Evaluasi berdasarkan classification report \n\n%s\n" % (classification_report(y_test, prediction,zero_division=0)))
    print()
    print("Confusion Matrix")
    print()
    y_actual = pd.Series(np.array(y_test), name = "actual")
    y_pred = pd.Series(np.array(prediction), name = "prediction")
    df_confusion = pd.crosstab(y_actual, y_pred)
    display(df_confusion)
    print()
    print()

    print('Accuracy Average:', accuracy_score(y_test, prediction))
    print('F1 Macro Average:', f1_score(y_test, prediction, average='macro'))
    print('F1 Micro Average:', f1_score(y_test, prediction, average='micro'))
    print('Precision Macro Average:', precision_score(y_test, prediction, average='macro',zero_division=0))
    print('Precision Micro Average:', precision_score(y_test, prediction, average='micro',zero_division=0))
    print('Recall Macro Average:', recall_score(y_test, prediction, average='macro',zero_division=0))
    print('Recall Micro Average:', recall_score(y_test, prediction, average='micro',zero_division=0))
    print()
```
